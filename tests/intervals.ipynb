```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor

# Semilla para reproducibilidad
np.random.seed(42)

# Crear serie sint√©tica
n_samples = 200
t = np.arange(n_samples)
y = np.sin(t / 20) + np.random.normal(0, 0.3, size=n_samples)

# Generar variables con rezagos (lags)
n_lags = 5
X = np.vstack([t])
for lag in range(1, n_lags + 1):
    lagged = np.roll(y, lag)
    X = np.vstack([X, lagged])
X = X.T
X[:n_lags, :] = np.nan

# Eliminar filas con NaNs
mask = ~np.isnan(X).any(axis=1)
X = X[mask]
y = y[mask]

# Separar entrenamiento y prueba
train_size = 100
cal_size = 50
X_train, X_cal,X_test = X[:train_size], X[train_size:train_size+cal_size],X[train_size+cal_size:]
y_train, y_cal ,y_test = y[:train_size], y[train_size:train_size+cal_size],y[train_size+cal_size:]

alpha =0.1
```
